```python
#Global Function
name = 'This is a global string'

#Enclosing Function Locals
def greet():
    name = 'Vallabh'
    def hello():
        #local 
        name = 'Local Name'
        print('Hello ' + name)

    hello()
```

```python
greet()
```

```python
x= 50

def func(x):
    print(f'X is {x}')

    #local Reassignment
    x = 200
    print(f'Changed X to : {x}')
```

```python
func(x)
```

```python
x= 50

def func():
    global x
    print(f'X is {x}')

    #local Reassignment on a Global Variable
    x = 'New Value'
    print(f'Changed Global X to : {x}')
```

```python
print(x)
```

```python
func()
```
